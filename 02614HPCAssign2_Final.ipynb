```python
df_original['Speedup'] = 0.0

# Iterate over each Grid_Size
for grid_size in df_original['Grid_Size'].unique():
    # Get the execution time for 1 thread (baseline)
    baseline_time = df_original[(df_original['Grid_Size'] == grid_size) & (df_original['Threads'] == 1)]['Execution_Time'].values[0]
    
    # Calculate speedup for each thread count
    df_original.loc[df_original['Grid_Size'] == grid_size, 'Speedup'] = baseline_time / df_original[df_original['Grid_Size'] == grid_size]['Execution_Time']
```

```python
df_optimized['Speedup'] = 0.0

# Calculate speedup for each grid size
for grid_size in df_optimized['Grid_Size'].unique():
    # Get the execution time for 1 thread (baseline)
    baseline_time = df_optimized[(df_optimized['Grid_Size'] == grid_size) & (df_optimized['Threads'] == 1)]['Execution_Time'].values[0]
    
    # Calculate speedup for each thread count
    df_optimized.loc[df_optimized['Grid_Size'] == grid_size, 'Speedup'] = baseline_time / df_optimized[df_optimized['Grid_Size'] == grid_size]['Execution_Time']
```

```python
# Create the plot with two subplots for Speedup and MFLOPS
fig, ax = plt.subplots(2, 2, figsize=(14, 10))

# Plot Speedup vs Threads for Original Data
for grid_size in df_original['Grid_Size'].unique():
    subset = df_original[df_original['Grid_Size'] == grid_size]
    ax[0, 0].plot(subset['Threads'], subset['Speedup'], marker='o', label=f"Grid Size {grid_size}")
ax[0, 0].set_title("Original Speedup vs Threads")
ax[0, 0].set_xlabel("Number of Threads")
ax[0, 0].set_ylabel("Speedup")
ax[0, 0].legend()
ax[0, 0].grid()

# Plot Speedup vs Threads for Optimized Data
for grid_size in df_optimized['Grid_Size'].unique():
    subset = df_optimized[df_optimized['Grid_Size'] == grid_size]
    ax[0, 1].plot(subset['Threads'], subset['Speedup'], marker='o', label=f"Grid Size {grid_size}")
ax[0, 1].set_title("Optimized Speedup vs Threads")
ax[0, 1].set_xlabel("Number of Threads")
ax[0, 1].set_ylabel("Speedup")
ax[0, 1].legend()
ax[0, 1].grid()

# Plot MFLOPS vs Threads for Original Data
for grid_size in df_original['Grid_Size'].unique():
    subset = df_original[df_original['Grid_Size'] == grid_size]
    ax[1, 0].plot(subset['Threads'], subset['MFLOPS'], marker='o', label=f"Grid Size {grid_size}")
ax[1, 0].set_title("Original MFLOPS vs Threads")
ax[1, 0].set_xlabel("Number of Threads")
ax[1, 0].set_ylabel("MFLOPS")
ax[1, 0].legend()
ax[1, 0].grid()

# Plot MFLOPS vs Threads for Optimized Data
for grid_size in df_optimized['Grid_Size'].unique():
    subset = df_optimized[df_optimized['Grid_Size'] == grid_size]
    ax[1, 1].plot(subset['Threads'], subset['MFLOPS'], marker='o', label=f"Grid Size {grid_size}")
ax[1, 1].set_title("Optimized MFLOPS vs Threads")
ax[1, 1].set_xlabel("Number of Threads")
ax[1, 1].set_ylabel("MFLOPS")
ax[1, 1].legend()
ax[1, 1].grid()
plt.savefig(f"speedup_grid.png")

# Display the plots
plt.tight_layout()
plt.show()
```
